```python
#Importando a biblioteca
import pandas as pd
```

```python
#Leitura dos arquivos
data_frame1 = pd.read_excel("/content/drive/MyDrive/datasets/Aracaju.xlsx")
data_frame2 = pd.read_excel("/content/drive/MyDrive/datasets/Fortaleza.xlsx")
data_frame3 = pd.read_excel("/content/drive/MyDrive/datasets/Natal.xlsx")
data_frame4 = pd.read_excel("/content/drive/MyDrive/datasets/Salvador.xlsx")
data_frame5 = pd.read_excel("/content/drive/MyDrive/datasets/Salvador.xlsx")
```

```python
#juntando todos os arquivos
data_frame = pd.concat([data_frame1,data_frame2,data_frame3,data_frame4,data_frame5])
```

# **VISUALIZAÇÃO DE DADOS**



```python
data_frame.head()
```

```python
data_frame["Receita"] = data_frame["Vendas"].mul(data_frame["Qtde"])
```

```python
data_frame["Ano_Venda"] = data_frame["Data"].dt.year
data_frame
```

```python
data_frame["Dia_Venda"] = data_frame["Data"].dt.day
data_frame
```

```python
#Exclui uma coluna:
data_frame.drop("mes_venda", axis = 1, inplace = True)
```

```python
data_frame.drop("Mes_Venda", axis = 1, inplace = True)
```

```python
data_frame.drop("Ano_Venda", axis = 1, inplace = True)
```

```python
data_frame["Mes_Venda"] = data_frame["Data"].dt.month
data_frame
```

```python
data_frame.sample(5)
```

```python
#Value_counts faz uma contage m de quantas linhas tem por ex: LojaID, ou seja 
#lojaID = 1036 foram contadas 185 linhas, logo, 1036 realizou 185 vendas
data_frame["LojaID"].value_counts(ascending = False)
```

```python
#Gráfico de barras
data_frame["LojaID"].value_counts(ascending = False).plot.bar()
```

```python
#Gráfico de barras horizontais
data_frame["LojaID"].value_counts(ascending = True).plot.barh()
```

```python
#Gráfico de Pizza
data_frame.groupby(data_frame["Data"].dt.year)["Receita"].sum().plot.pie()
```

```python
#Total vendas por cidade
data_frame["Cidade"].value_counts()
```

```python
#Adicionando um título e alterando o nome dos eixos, tenho que importar a biblioteca
#matplotlib.pyplot para configurar o gráfico
import matplotlib.pyplot as plt
data_frame["Cidade"].value_counts().plot.bar(title = "Total de vendas por Cidade")
plt.xlabel("Cidade")#plt é a biblioteca do gráfico precisamos importar ela para mexer nele
plt.ylabel("Total de Vendas");#plt é a biblioteca do gráfico precisamos importar ela para mexer nele
```

```python
#Alterando a cor
data_frame["Cidade"].value_counts().plot.bar(title = "Total de Vendas por Cidade", color = "purple")
plt.xlabel("Cidade") #plt é a biblioteca do gráfico precisamos importar ela para mexer nele
plt.ylabel("Total de Vendas")#plt é a biblioteca do gráfico precisamos importar ela para mexer nele
```

```python
#Alterando a cor 
plt.style.use("ggplot") #tem na documentação da biblioteca matplotlib.ptplot
```

```python
data_frame.groupby(data_frame["Mes_Venda"])["Qtde"].sum().plot(title = "Total de Produtos Vendidos por Mês")
plt.xlabel("Mês")
plt.ylabel("Total de Produtos Vendidos");
plt.legend()
```

```python
data_frame.groupby(data_frame["Mes_Venda"])["Qtde"].sum()
```

```python
#Selecionando apenas as vendas de 2019
data_frame_2019 = data_frame[data_frame["Ano_Venda"] == 2019]
```

```python
data_frame_2019.groupby(data_frame_2019["Mes_Venda"])["Qtde"].sum()
```

```python
#Total produtos vendidos por mês em 2019
data_frame_2019.groupby(data_frame_2019["Mes_Venda"])["Qtde"].sum().plot(marker = "v")
plt.xlabel("Mês")
plt.ylabel("Total de Produtos Vendidos");
plt.legend()
```
