# **Trabalhando com Planilhas do Excel**

```python
#Importando a biblioteca
import pandas as pd
```

```python
#Leitura dos arquivos
data_frame1 = pd.read_excel("/content/drive/MyDrive/datasets/Aracaju.xlsx")
data_frame2 = pd.read_excel("/content/drive/MyDrive/datasets/Fortaleza.xlsx")
data_frame3 = pd.read_excel("/content/drive/MyDrive/datasets/Natal.xlsx")
data_frame4 = pd.read_excel("/content/drive/MyDrive/datasets/Salvador.xlsx")
data_frame5 = pd.read_excel("/content/drive/MyDrive/datasets/Salvador.xlsx")
```

```python
data_frame1.head()
```

```python
#juntando todos os arquivos
data_frame = pd.concat([data_frame1,data_frame2,data_frame3,data_frame4,data_frame5])
```

```python
#Exibindo as primeiras 5 linhas
data_frame.head()
```

```python
#Exibindo as ultimas 5 linhas
data_frame.tail()
```

```python
#Amostra de 5 linhas aleátorias
data_frame.sample(5)
```

```python
#Tipos de dado de cada coluna
data_frame.dtypes
```

```python
#Alterando o tipo de dado da coluna LojaID: astype = como tipo string
data_frame["LojaID"] = data_frame["LojaID"].astype("object")
```

```python
data_frame.dtypes
```

### **TRATANDO VALORES FALTANTES**

```python
#Consultando linhas com valores faltantes
data_frame.isnull().sum()
```

```python
#Substituindo os valores nulos pela média:
data_frame["Vendas"].fillna(data_frame["Vendas"].mean(), inplace = True)
#Mostrando os valores nulos existentes:
data_frame.isnull().sum()
```

```python
#Apagando as linhas com valores nulos:
data_frame.dropna(inplace = True)
```

```python
data_frame.head()
```

```python
data_frame.sample(5)
```

```python
#Criando coluna de receita, .MUL de multiplicação
data_frame["Receita"] = data_frame["Vendas"].mul(data_frame["Qtde"])
data_frame.head()
```

```python
data_frame["Receita/Vendas"] = data_frame["Receita"] / data_frame["Vendas"]
```

```python
data_frame.head()
```

```python
#Retorna a maior receita
data_frame["Receita"].max()
```

```python
#Retorna a menor receita
data_frame["Receita"].min()
```

```python
#nlargest -> Retorna o top 3 com base na coluna receita, 3 LINHAS COM MAIORES RECEITAS
data_frame.nlargest(3,"Receita")
```

```python
#nsamllest - > Retorna o top 3 piores em receitas
data_frame.nsmallest(3, "Receita")
```

```python
#Agrupando por cidade, som,a das receitas em cada cidade
data_frame.groupby("Cidade")["Receita"].sum()
```

```python
#Ordenando o conjunto de dados, TOP 5 MELHORES RECEITAS 
data_frame.sort_values("Receita", ascending = False).head(5)
```

# Trabalhando com Datas

```python
#Transformando a coluna de data em tipo inteiro
data_frame["Data"] = data_frame["Data"].astype("int64")
data_frame.dtypes
```

```python
#Verificando o tipo dado de cada coluna
data_frame.dtypes
```

```python
#Transformando a coluna de data em data
data_frame["Data"] = pd.to_datetime(data_frame["Data"])
data_frame.dtypes
```

```python
#Agrupamento por ano, retorna a soma de todas as receitas nas datas 2018 e 2019
data_frame.groupby(data_frame["Data"].dt.year)["Receita"].sum()
```

```python
#Criando uma nova coluna com o ano
data_frame["Ano_Venda"] = data_frame["Data"].dt.year
```

```python
data_frame.sample(5)
```

```python
 #Extraindo o mês
 data_frame["mes_venda"] = data_frame["Data"].dt.month
 #Extraindo o dia
 data_frame["dia_mes"] = data_frame["Data"].dt.day
```

```python
data_frame.sample(5)
```

```python
#Retornando a data mais antiga
data_frame["Data"].min()
```

```python
#Calculando a diferença de dias
data_frame["diferenca_dias"] = data_frame["Data"] - data_frame["Data"].min()
data_frame.sample(5)
```

```python
#Criando a coluna de trimestre
data_frame["trimestre_venda"] = data_frame["Data"].dt.quarter
data_frame.sample(5)
```

```python
#Filtrando as vendas de 2019 do mês março, usando .loc[]
vendas_marco_19 = data_frame.loc[(data_frame["Data"].dt.year == 2019) & (data_frame["Data"].dt.month == 3)]
```

```python
vendas_marco_19.sample(5)
```
