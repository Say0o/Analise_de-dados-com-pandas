```python
#Para criar um dicionário utilizamos as {}
dicionario = {"Maça":20,"Banana":10, "Laranja":5}
dicionario
```

```python
dicionario["Maça"] = 25
dicionario
```

```python
#retorna chaves do dicionario
dicionario.keys()
```

```python
#Retorna os valores do dicionário
dicionario.values()
```

```python
#Verificando se já existe uma chave no dicionário e caso não exista inserir
dicionario.setdefault("Limão", 22)
dicionario
```
